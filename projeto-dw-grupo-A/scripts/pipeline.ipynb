# Conectar ao DuckDB, registrar os DataFrames (`df_customers`, `df_orders`, `df_order_items`, `df_sellers`, `df_products`) como views temporrias, e criar um data warehouse no DuckDB, que consiste em: uma tabela dimensional de data (`dim_date`), tabelas dimensionais de cliente (`dim_customer`, com suporte a SCD Tipo 2), produto (`dim_product`) e vendedor (`dim_seller`), e uma tabela fato de vendas (`fact_sales`).

```python
import duckdb

# Create an in-memory DuckDB connection
con = duckdb.connect(database=':memory:', read_only=False)

# Register DataFrames as temporary views
con.register('df_customers_view', df_customers)
con.register('df_orders_view', df_orders)
con.register('df_order_items_view', df_order_items)
con.register('df_sellers_view', df_sellers)
con.register('df_products_view', df_products)

print("DataFrames registered as temporary views in DuckDB successfully.")
```

```python
con.execute("""
--DROP TABLE IF EXISTS dim_date;
CREATE TABLE dim_date (
    date_sk INTEGER PRIMARY KEY,
    full_date DATE,
    year INTEGER,
    month INTEGER,
    day INTEGER,
    day_of_week INTEGER,
    month_name VARCHAR,
    day_name VARCHAR
);
""")
print("Table 'dim_date' created or recreated successfully.")
```

```python
con.view('dim_date')
```

```python
con.execute("""
INSERT INTO dim_date (date_sk, full_date, year, month, day, day_of_week, month_name, day_name)
SELECT
    ROW_NUMBER() OVER (ORDER BY full_date_clean) AS date_sk,
    full_date_clean,
    EXTRACT(YEAR FROM full_date_clean) AS year,
    EXTRACT(MONTH FROM full_date_clean) AS month,
    EXTRACT(DAY FROM full_date_clean) AS day,
    -- DuckDB's EXTRACT(DOW FROM date) returns 0 for Sunday, 1 for Monday...6 for Saturday.
    -- To match the requirement (1=Domingo, 2=Segunda, etc.), we add 1.
    EXTRACT(DOW FROM full_date_clean) + 1 AS day_of_week,
    strftime(full_date_clean, '%B') AS month_name,
    strftime(full_date_clean, '%A') AS day_name
FROM (
    SELECT DISTINCT DATE_TRUNC('day', CAST(order_purchase_timestamp AS TIMESTAMP)) AS full_date_clean
    FROM df_orders_view
) AS unique_dates;
""")

print("Table 'dim_date' populated successfully.")

# Verify the first 5 rows
result = con.execute("SELECT * FROM dim_date LIMIT 5;").fetchdf()
print("First 5 rows of dim_date:")
print(result)
```

```python
con.view('dim_date')
```

```python
con.execute("""
CREATE TABLE dim_customer (
    customer_id VARCHAR PRIMARY KEY,
    customer_unique_id VARCHAR,
    customer_zip_code_prefix INTEGER,
    customer_city VARCHAR,
    customer_state VARCHAR
);
""")
print("Table 'dim_customer' created successfully.")
```

```python
# populate dim_customer
con.execute("""
INSERT INTO dim_customer (customer_id, customer_unique_id,  customer_zip_code_prefix, customer_city, customer_state)
SELECT
    customer_id,  -- Assuming customer_id is a unique identifier
    customer_unique_id,
    customer_zip_code_prefix,
    customer_city,
    customer_state
FROM df_customers_view;
""")
```

```python
con.view('dim_customer')
```

```python
# create dim_seller

con.execute("""
CREATE TABLE dim_seller (
    seller_id VARCHAR PRIMARY KEY,
    seller_zip_code_prefix INTEGER,
    seller_city VARCHAR,
    seller_state VARCHAR
);
""")
print("Table 'dim_seller' created successfully.")
```

```python
con.view('dim_seller')
```

```python
# populate dim_seller
con.execute("""
INSERT INTO dim_seller (seller_id, seller_zip_code_prefix, seller_city, seller_state)
SELECT
    seller_id,
    seller_zip_code_prefix,
    seller_city,
    seller_state
FROM df_sellers_view;
""")

print("Table 'dim_seller' populated successfully.")
```

```python
con.view('dim_seller')
```

```python
con.execute("""
DROP TABLE IF EXISTS dim_product;
CREATE TABLE dim_product (
    product_id VARCHAR PRIMARY KEY,
    product_category_name VARCHAR,
    product_name_lenght INTEGER,
    product_description_lenght INTEGER,
    product_photos_qty INTEGER,
    product_weight_g INTEGER,
    product_length_cm INTEGER,
    product_height_cm INTEGER,
    product_width_cm INTEGER
);
""")
print("Table 'dim_product' created successfully.")

# populate dim_product
con.execute("""
INSERT INTO dim_product (
    product_id,
    product_category_name,
    product_name_lenght,
    product_description_lenght,
    product_photos_qty,
    product_weight_g,
    product_length_cm,
    product_height_cm,
    product_width_cm
)
SELECT
    product_id,
    product_category_name,
    product_name_lenght,
    product_description_lenght,
    product_photos_qty,
    product_weight_g,
    product_length_cm,
    product_height_cm,
    product_width_cm
FROM df_products_view;
""")

print("Table 'dim_product' populated successfully.")

# Verify the first 5 rows
result = con.execute("SELECT * FROM dim_product LIMIT 5;").fetchdf()
print("First 5 rows of dim_product:")
print(result)
```

```python
con.view('dim_product')
```

```python
con.execute("""
DROP TABLE IF EXISTS fact_sales; -- Drop the table to apply schema changes
CREATE TABLE fact_sales (
    order_id VARCHAR,
    customer_id VARCHAR,
    order_purchase_timestamp TIMESTAMP,
    order_item INTEGER,
    product_id VARCHAR,
    seller_id VARCHAR,
    shipping_limit_date TIMESTAMP,
    price FLOAT,
    freight_value FLOAT,
    PRIMARY KEY (order_id, order_item), -- Composite primary key
    FOREIGN KEY (customer_id) REFERENCES dim_customer(customer_id)
    -- FOREIGN KEY (product_id) REFERENCES dim_product(product_id),
    -- FOREIGN KEY (seller_id) REFERENCES dim_seller(seller_id)
);
""")
print("Table 'fact_sales' recreated successfully with composite primary key.")
```

```python
con.view('fact_sales')
```

```python
# popular fact_sales

con.execute("""
INSERT INTO fact_sales (order_id, customer_id, order_purchase_timestamp, order_item, product_id, seller_id, shipping_limit_date, price, freight_value)
SELECT
    o.order_id,
    o.customer_id,
    o.order_purchase_timestamp,
    oi.order_item_id,
    oi.product_id,
    oi.seller_id,
    oi.shipping_limit_date,
    oi.price,
    oi.freight_value
FROM df_orders_view o
JOIN df_order_items_view oi ON o.order_id = oi.order_id;
""")

print("Table 'fact_sales' populated successfully.")

result = con.execute("SELECT * FROM fact_sales LIMIT 5;").fetchdf()
print("First 5 rows of fact_sales:")
print(result)
```

Viso Geral do Esquema Estrela (Star Schema)
O data warehouse consiste em uma Tabela Fato central (fact_sales) cercada por quatro Tabelas Dimenso (dim_date, dim_customer, dim_product, dim_seller).

Tabela Fato: fact_sales
Propsito: Armazena detalhes individuais das transaes de vendas e medidas.
Chave Primria: Chave composta: (order_id, order_item)
Colunas:

order_id: Identificador do pedido.
customer_id: Chave estrangeira referenciando dim_customer.
order_purchase_timestamp: Data e hora da compra do pedido.
order_item: Nmero do item dentro do pedido.
product_id: Chave estrangeira referenciando dim_product.
seller_id: Chave estrangeira referenciando dim_seller.
shipping_limit_date: Data limite para envio do item.
price: Preo do item.
freight_value: Custo do frete do item.

Relacionamentos (Chaves Estrangeiras):

customer_id  dim_customer.customer_id
product_id  dim_product.product_id
seller_id  dim_seller.seller_id
order_purchase_timestamp (parte da data)  dim_date.date_sk


Tabela Dimenso: dim_date
Propsito: Fornece atributos temporais para anlise, permitindo filtragem e agregao por vrios componentes de data.
Chave Primria: date_sk (chave substituta inteira)
Colunas:

date_sk: Chave substituta para a data.
full_date: Data completa.
year: Ano da data.
month: Ms da data.
day: Dia do ms.
day_of_week: Dia da semana (1=Domingo, 7=Sbado).
month_name: Nome completo do ms.
day_name: Nome completo do dia da semana.

Relacionamento com fact_sales:
fact_sales.order_purchase_timestamp est vinculado a dim_date.full_date (ou date_sk).

Tabela Dimenso: dim_customer
Propsito: Contm informaes sobre os clientes.
Chave Primria: customer_id
Colunas:

customer_id: Identificador nico do cliente.
customer_unique_id: Identificador nico do cliente em mltiplas compras.
customer_zip_code_prefix: Primeiros cinco dgitos do CEP do cliente.
customer_city: Cidade do cliente.
customer_state: Estado do cliente.

Relacionamento com fact_sales:
fact_sales.customer_id  dim_customer.customer_id

Tabela Dimenso: dim_product
Propsito: Contm dados descritivos sobre os produtos.
Chave Primria: product_id
Colunas:

product_id: Identificador nico do produto.
product_category_name: Categoria do produto.
product_name_length: Comprimento do nome do produto.
product_description_length: Comprimento da descrio do produto.
product_photos_qty: Quantidade de fotos do produto.
product_weight_g: Peso do produto em gramas.
product_length_cm: Comprimento do produto em centmetros.
product_height_cm: Altura do produto em centmetros.
product_width_cm: Largura do produto em centmetros.

Relacionamento com fact_sales:
fact_sales.product_id  dim_product.product_id

Tabela Dimenso: dim_seller
Propsito: Contm informaes sobre os vendedores.
Chave Primria: seller_id
Colunas:

seller_id: Identificador nico do vendedor.
seller_zip_code_prefix: Primeiros cinco dgitos do CEP do vendedor.
seller_city: Cidade do vendedor.
seller_state: Estado do vendedor.

Relacionamento com fact_sales:
fact_sales.seller_id  dim_seller.seller_id
